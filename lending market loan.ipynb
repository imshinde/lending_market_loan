## Handling missing values

```python
lending_df.isnull().sum()
```

```python
lending_df.columns
```

```python
for col in lending_df:
    print(col,":",lending_df[col].isnull().sum(), "null value",(lending_df[col].isnull().sum()/lending_df.shape[0])*100,"%")
```

```python
lending_df.shape
```

```python
null_cols = []
for col in lending_df:
    if (lending_df[col].isnull().sum()/lending_df.shape[0])*100 > 50:
        null_cols.append(col)
```

```python
print(null_cols)
```

```python
lending_df.drop(columns= null_cols,axis=1,inplace= True)
```

```python
lending_df.shape
```

```python
for col in lending_df:
    print(col,":",lending_df[col].isnull().sum(), "null value")
```

```python
lending_df['desc']
```

```python
lending_df['desc'].fillna('not describe',inplace=True)
```

```python
for col in lending_df:
    print(col,":",lending_df[col].isnull().sum(), "null value",(lending_df[col].isnull().sum()/lending_df.shape[0])*100,"%")
```

### First 20 Columns

```python
lending_df[cols[:20]]
```

### funded_amnt_inv

```python
lending_df['funded_amnt_inv'] = lending_df['funded_amnt_inv'].apply(lambda x:np.round(x, decimals=2))
```

### term

```python
lending_df['term'].value_counts()
```

```python
import re
```

```python
lending_df['term'] = lending_df['term'].apply(lambda x:re.findall(r'[\d.]+', x)[0])
```

### int_rate

```python
lending_df['int_rate'] = lending_df['int_rate'].apply(lambda x:re.findall(r'[\d.]+', x)[0])
```

### emp_length

```python
lending_df['emp_length'] = lending_df['emp_length'].replace({'< 1 year':'0'})
```

```python
lending_df['emp_length'] = lending_df['emp_length'].apply(lambda x:re.findall(r'[\d.]+', x)[0])
```

### revol_util

```python
lending_df['revol_util'] = lending_df['revol_util'].apply(lambda x:re.findall(r'[\d.]+', x)[0])
```

### total_pymnt

```python
lending_df['total_pymnt'] = lending_df['total_pymnt'].apply(lambda x:np.round(x, decimals=2))
```
