# Data Analysis & Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lending_df = pd.read_csv('final_loans_data.csv') 
```

## Total Grade

```python
lending_df['grade'].value_counts()
```

```python
sns.countplot(x='grade', data = lending_df)
```

## Sub Grade

```python
plt.figure(figsize=(10,16))
sns.countplot(y='sub_grade', data = lending_df)
```

## Term

```python
sns.countplot(x='term', data = lending_df)
```

## Emp length


```python
plt.figure(figsize=(8,12))
sns.countplot(x='emp_length', data = lending_df)
```

## Home ownership

```python
sns.countplot(x='home_ownership', data = lending_df)
```

## verification_status

```python
sns.countplot(x='verification_status', data = lending_df)
```

```python
lending_df['verification_status'].value_counts()
```

```python
plt.figure(figsize=(10,8))
plt.pie(lending_df['verification_status'].value_counts(),autopct='%.0f%%',labels=['Not Verified','Verified','Source Verified'])
plt.show()
```

## Loan status

```python
lending_df['loan_status'].value_counts()
```

```python
sns.countplot(y='loan_status',data = lending_df)
```

```python
plt.figure(figsize=(10,8))
plt.pie(lending_df['loan_status'].value_counts(),autopct='%.0f%%',labels=['fully paid','charged off','current','','','','','','',])
plt.show()
```

## purpose

```python
sns.countplot(y='purpose',data = lending_df)
```

##  annual income vs loan ammount

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x = 'annual_inc', y = 'loan_amnt', hue = 'grade',data = lending_df)
plt.title('annual income vs loan ammount')
```

## Converting Datetime

```python
lending_df['issue_d'] = pd.to_datetime(lending_df['issue_d'])
lending_df['earliest_cr_line'] = pd.to_datetime(lending_df['earliest_cr_line'])
lending_df['last_pymnt_d'] = pd.to_datetime(lending_df['last_pymnt_d'])
lending_df['last_credit_pull_d'] = pd.to_datetime(lending_df['last_credit_pull_d'])
```

## Loan Amount Forecasting

```python
plt.figure(figsize=(14,8))
sns.lineplot(x='issue_d', y='loan_amnt', data=lending_df)
plt.title('Loan Amount Forecasting')
plt.show()
```
